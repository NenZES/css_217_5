```python
def chooseOption(options):
    print("Choose an option:")
    for i, option in enumerate(options, start=1):
        print(f"{i}. {option}")
    choice = int(input("Enter your choice: "))
    return choice

def customizeCharacter(character):
    print("\nCustomize Appearance:")
    hairstyle_options = ["Long Hair", "Short Hair", "Medium Hair", "Bald"]
    body_type_options = ["Slim", "Muscular", "Athletic"]
    skin_tone_options = ["Fair", "Pale", "Tan"]

    character.appearance.hairstyle = hairstyle_options[chooseOption(hairstyle_options) - 1]
    character.appearance.body_type = body_type_options[chooseOption(body_type_options) - 1]
    character.appearance.skin_tone = skin_tone_options[chooseOption(skin_tone_options) - 1]

    print("\nCustomize Abilities:")
    abilities = []
    print("Choose abilities (enter numbers separated by commas):")
    ability_options = ["Slash: Powerful melee attack", "Block: Defensive stance", "Fireball: Launches a ball of fire", "Teleport: Instantly move to a location", "Arrow Shot: Shoots arrows with precision", "Dodge: Quickly evade attacks"]
    print("\nAvailable abilities:")
    for i, ability in enumerate(ability_options, start=1):
        print(f"{i}. {ability}")
    choices = input("Enter your choices (e.g. 1, 3, 5): ").split(",")
    for choice in choices:
        abilities.append(Ability(*ability_options[int(choice) - 1].split(": ")))
    character.abilities = abilities

    print("\nCustomize Equipment:")
    equipment = []
    print("Choose equipment (enter numbers separated by commas):")
    equipment_options = ["Sword: A sturdy blade", "Shield: Protects against attacks", "Staff: Focuses magical energy", "Robe: Enhances magical abilities", "Bow: Long-range weapon", "Quiver: Holds arrows"]
    print("\nAvailable equipment:")
    for i, equip in enumerate(equipment_options, start=1):
        print(f"{i}. {equip}")
    choices = input("Enter your choices (e.g. 1, 3, 5): ").split(",")
    for choice in choices:
        equipment.append(Equipment(*equipment_options[int(choice) - 1].split(": ")))
    character.equipment = equipment

    print("\nCustomization completed successfully!")
    return character


def main():
    print("Welcome to Character Creator!")
    print("Choose a class:")
    print("1. Warrior")
    print("2. Mage")
    print("3. Archer")

    class_choice = int(input("Enter your choice: "))

    if class_choice == 1:
        factory = WarriorFactory()
    elif class_choice == 2:
        factory = MageFactory()
    elif class_choice == 3:
        factory = ArcherFactory()
    else:
        print("Invalid choice.")
        return

    creator = CharacterCreator()
    creator.setFactory(factory)

    name = input("Enter your character's name: ")
    character = creator.createCharacter(name)

    if character:
        print("\nCharacter created successfully!")
        print("Name:", character.name)
        print("Appearance:", character.appearance.__dict__)
        print("Abilities:")
        for ability in character.abilities:
            print("- {} : {}".format(ability.name, ability.description))
        print("Equipment:")
        for equipment in character.equipment:
            print("- {} : {}".format(equipment.name, equipment.description))
        print("Attributes:", character.attributes.__dict__)

        customize = input("\nDo you want to customize your character? (yes/no): ")
        if customize.lower() == "yes":
            customized_character = customizeCharacter(character)
            print("\nUpdated Character Details:")
            print("Name:", customized_character.name)
            print("Appearance:", customized_character.appearance.__dict__)
            print("Abilities:")
            for ability in customized_character.abilities:
                print("- {} : {}".format(ability.name, ability.description))
            print("Equipment:")
            for equipment in customized_character.equipment:
                print("- {} : {}".format(equipment.name, equipment.description))
            print("Attributes:", customized_character.attributes.__dict__)

    else:
        print("Failed to create character.")

if __name__ == "__main__":
    main()
```
